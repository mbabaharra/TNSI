<h1>Chapitre 1 : Initiation au langage Python : les éléments d'un programme</h1>

Un **algorithme**  est une suite `finie` et `non ambiguë` d'instructions permettant de résoudre un problème.

Un **programme**  est une traduction d'un algorithme écrit dans un langage de programmation. C'est-à-dire un langage *compris* et *exécuté* par une machine. La notion d'algorithme prend une partie importante de votre programme de première et de terminale NSI.

Il existe deux grandes familles de langages de programmation :
1. Les langages de **bas niveau** : Le langage **assembleur**, qui est un langage *proche de la machine*. Ils sont très difficile  à utiliser, car très  éloignés du langage naturel. voir <a href ="https://www.ninjaone.com/fr/it-hub/remote-access/qu-est-ce-qu-un-langage-de-bas-niveau/">Qu’est-ce qu’un langage de bas niveau ? </a>

2. Les langages de haut niveau : ils sont plus proches du langage naturel, donc plus "faciles" à utiliser,  
voir <a href ="https://www.ninjaone.com/fr/it-hub/it-service-management/qu-est-ce-qu-un-langage-de-programmation-de-haut-niveau/" >Qu’est-ce qu’un langage de programmation de haut niveau ?  </a>

Pour les deux langage de programmation, un programme est une suite de taches (**instructions**) « traduites » en langage machine grâce à un programme spécialisé appelé **interpréteur** ou bien **compilateur**, suivant la méthode utilisée pour effectuer la traduction

<h2><span style='color:red'>C- Mémoriser des données : les variables</span></h2>
Le concept de variable est très important en programmation. <br/>

Un programme *passe* son temps à manipuler des données. Il doit ranger  donc ces données dans la mémoire (RAM - Random Access Memory) de l'ordinateur (une donnée dans une case).
une variable est une information (une donnée) temporaire que l'on stocke dans une case nommée de la RAM. On dit qu'elle est "variable", car c'est une valeur qui peut changer pendant le déroulement du programme. 
Une variable est l'association de  :
- une valeur présente en mémoire (par exemple le nombre entier 1992)
- un nom (Par exemple. annee_naissance = 1992). 

Cette association "nom = valeur" est appelée affectation. 
Le signe `=` est ici un signe d'<b>affectation</b> qui permet donc d'attribuer une valeur (ou autre) à notre variable

<h3><span style='color:red'>B- Tests et conditions</span></h3>
<h4><span style='color:green'>a- Instruction if</span></h4>
La structure conditionnelle permettant d'effectuer un test s'écrit en utilisant l'expression <b>if</b>.<br/>

````python
# Tester le code ci-dessous
age = 9
if age < 10:
    print("Accès interdit")
print("Fin")
````

On constate ici que lorsque $age$ contient l'entier 9, la condition booléenne $age < 10$ (age < 10 est une expression booléenne) est vraie donc la ligne 4 est exécutée. Une fois que cette ligne 4 est exécutée, le script se poursuit, on "sort du bloc du if" et le script se poursuit à la ligne 5.

Attention :
* le : (deux-points) à la fin de la ligne 3 est obligatoire pour respecter la syntaxe Python.
* L'indentation (décalage vers le droite d'une tabulation) marque les lignes qui composent le bloc du if. Ici ce bloc du if est composé de la seule ligne 4.
* La ligne 5 n'est plus indentée, ce qui signale qu'elle ne fait plus partie du bloc du if.

Exécutez la *cellules Code* suivante  :

````python
age = 11
if age < 10:
    print("Accès interdit")
print("Fin")
````

Ici pour la valeur de 11, la condition booléenne $age < 10$ de la ligne 2 est fausse donc le bloc du if n'est pas exécuté. Autrement dit on passe directement à la ligne 4 sans exécuter la ligne 3.

<h4><span style='color:green'>b- Instruction if...else</span></h4>
Exécutez le script suivant :

````python
age = 9
if age < 10:
    print("Accès interdit")
else:
    print("Accès autorisé")
print("Fin")
````

Lorsque $age$ contient l'entier 9, la condition booléenne $age < 10$ est vraie donc la ligne 3 est exécutée. Une fois cette ligne 3 est exécutée, le script se poursuit, on "sort du if" et le else n'est pas exécuté, le script se poursuit à la ligne 6.

Dans la ligne 4, le else est suivi d'un : (deux-points) sans condition booléenne.

Exécutez le script suivant :

````python
age = 11
if age < 10:
    print("Accès interdit")
else:
    print("Accès autorisé")
print("Fin")
````

Lorsque $age$ contient l'entier 11, la condition booléenne $age < 10$ est fausse donc la ligne 3 n'est pas exécutée. On passe alors au else, la ligne 5 est exécutée, et le script se termine par la ligne 6.

A noter que :
* elif est obligatoirement suivi d'une condition booléenne et se termine par un : (deux-points).
* On n'est forcément dans un des trois cas : $x<10$ (cas du if) ou $x==0$ (cas du elif) ou $x>10$ (cas du else). Lorsqu'un de ces trois booléens est vrai, on passe directement à la ligne 8 puisque les autres situations ne peuvent pas arriver.

Le script suivant semble donner le même résultat, pourtant il est moins performant que le précédent car chacune des lignes 2, 4 et 6 est toujours exécutée.

````python
x=110
if x<0:
    print("Le nombre est strictement négatif")
if x==0:
    print("Le nombre est nul")
if x>0:
    print("Le nombre est strictement positif")
print("Fin")
````
